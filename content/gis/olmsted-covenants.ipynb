```python
# Ensure that the necessary packages are installed
# Load Geopandas
import geopandas as gpd
import pyodide_http
from folium.plugins import Fullscreen, MousePosition

%pip install -qq geopandas folium mapclassify
```

```python
pyodide_http.patch_all()
```

```python
# Load the Mapping Prejudice data from Github
olmsted_shapefile = "https://raw.githubusercontent.com/UMNLibraries/mp-us-racial-covenants/main/mn-olmsted-county/covenants-mn-olmsted-county.zip"
```

```python
# Load covenants into a GeoPandas dataframe
olmsted_covenants = gpd.read_file(olmsted_shapefile)
olmsted_covenants.head()
```

```python
# Map the covenants on a Folium map
map = olmsted_covenants.explore()
Fullscreen(
    position="topright",
    title="Expand me",
    title_cancel="Exit me",
    force_separate_button=True,
).add_to(map)
MousePosition().add_to(map)

map
```
