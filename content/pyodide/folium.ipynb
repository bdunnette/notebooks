## Demo of `folium` Map

Load in the `folium` package:

```python
import folium
from folium.plugins import MousePosition, Fullscreen
```

And render a demo map:

```python
m = folium.Map(location=[44.9778, -93.2650], zoom_start=12)

MousePosition().add_to(m)

Fullscreen(
    position="topright",
    title="Fullscreen",
    title_cancel="Exit fullscreen",
    force_separate_button=True,
).add_to(m)

m
```
